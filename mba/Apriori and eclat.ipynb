```python
from apyori import apriori
```

```python
len(dataset)
```

```python
transactions = []
for i in range(0, len(dataset)):
    transactions.append([  str(dataset.values[i,j])
                         for j in range(0,len(dataset.loc[i]))])
```

```python
transactions[0]
```

```python
rules = apriori(transactions = transactions, 
                min_support=0.003, 
               min_confidence=0.2,
               min_lift=3,
                min_length=3,
                max_length=3
               )
```

```python
results = list(rules)
```

```python
results
```

```python
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    supports = [result[1] for result in results]
    confidence = [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports,  confidence, lifts))
    
```

```python
rindf = pd.DataFrame(inspect(results), columns=['left hand side', 'right hand side', 'support', 'c', 'l'])
```

```python
rindf
```

```python
rindf.nlargest(n=10, columns='lift')
```
